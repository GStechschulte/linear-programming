```python
from linopy import Model
from scipy.optimize import linprog
```

```python
m = Model()

x1 = m.add_variables(lower=0, upper=6000, name="x1")
x2 = m.add_variables(lower=0, upper=4000, name="x2")

m.add_constraints(((1/200 * x1) + (1/140 * x2)) <= 40)

m.add_objective(-25 * x1 + -30 * x2)
```

```python
m.solve()
```

```python
x1.solution
```

```python
x2.solution
```

```python
obj_func = [-25, -30]
lhs = [
    [1/200, 1/140],
    [1, 0],
    [0, 1]
    ]
rhs = [40, 6000, 4000]

opt = linprog(obj_func, A_ub=lhs, b_ub=rhs, method="highs")
```
